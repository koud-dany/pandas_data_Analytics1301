```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merged_df = pd.merge(bitcoin_df, dash_df, on='Date', suffixes=('_bitcoin', '_dash'))
merged_df.head()
```

```python
# Rename columns so that they are differentiated
```

```python
# alternatively you can set your suffixes when the merge occurs
```

```python
# Collecting best open for Bitcoin and Dash
bitcoin_open = merged_df['Open_bitcoin'].max()
dash_open = merged_df['Open_dash'].max()
# Collecting best close for Bitcoin and Dash
bitcoin_close = merged_df['Close_bitcoin'].max()
dash_close = merged_df['Close_dash'].max()

# Collecting the total volume for Bitcoin and Dash
bitcoin_vol = merged_df['Volume_bitcoin'].sum()
dash_vol = merged_df['Volume_dash'].sum()
```

```python
# Creating a summary DataFrame using above values
#Best Bitcoin Open, Best Dash Open, Best Bitcoin Close, Best Dash Close, Total Bitcoin Volume, Total Dash Volume
pd.DataFrame({
    'Best Bitcoin Open': [bitcoin_open],
    'Best Dash Open': [dash_open],
    'Best Bitcoin Close': [bitcoin_close],
    'Best Dash Close': [dash_close],
    'Total Bitcoin Volume in Millions': [round(bitcoin_vol/1000000, 2)],
    'Total Dash Volume in Millions': [round(dash_vol/1000000, 2)]
})
```
