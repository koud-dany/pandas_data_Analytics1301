```python
import pandas as pd
import os
```

```python
# The path to our CSV file
file_path = os.path.join('Resources','KickstarterData.csv')
# Read our Kickstarter data into pandas
ks_df = pd.read_csv(file_path)
ks_df.head()
```

```python
# Get a list of all of our columns for easy reference
ks_df.columns
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
ks_df = ks_df[["name", "goal", "pledged", "state", "country", "staff_pick","backers_count", "spotlight"]]
ks_df.head()
```

```python
ks_df.info()
```

```python
# Remove projects that made no money at all
ks_df = ks_df.loc[ks_df['pledged'] > 0]
ks_df.info()
```

```python
# Collect only those projects that were hosted in the US
# Create a list of the columns
# Create a new df for "US" with the columns above. 
ks_us_df = ks_df.loc[ks_df['country'] == 'US']
ks_us_df.info()
```

```python
# Create a new column that finds the average amount pledged to a project
ks_us_df.loc[:, 'avg_per_backer'] = ks_us_df['pledged'] / ks_us_df['backers_count']
```

```python
ks_us_df.head()
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float
# Then Format to go to two decimal places, include a dollar sign, and use comma notation
ks_us_df.info()
```

```python
ks_us_df['avg_per_backer'] = ks_us_df['avg_per_backer'].map('${:,.2f}'.format)
ks_us_df.head()
```

```python
# ks_us_df['goal'] = ks_us_df['goal'].astype('float')
```

```python
# Calculate the total number of backers for all US projects
ks_us_df['backers_count'].sum()
```

```python
# Calculate the average number of backers for all US projects
ks_us_df['backers_count'].mean()
```

```python
# Collect only those US campaigns that have been picked as a "Staff Pick"
ks_us_staffpicks_df = ks_us_df.loc[ks_us_df['staff_pick']]
```

```python
# Group by the state of the campaigns and see if staff picks matter
ks_us_df.groupby(['state','staff_pick'])['name'].count()
```
