```python
# Import Dependencies
import pandas as pd
```

```python
raw_data = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [90, 59, 72, 88, 98, 60]}
df = pd.DataFrame(raw_data)
df
```

```python
raw_data_2 = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [0, 59.9, 59.91, 69.9, 100.1, 100]}
df2 = pd.DataFrame(raw_data_2)
df2
```

```python
# Create the bins in which Data will be held
# The cutting points for the bins are 0, 59.9, 69.9, 79.9, 89.9, 100.   
bins = [0, 59.9, 69.9, 79.9, 89.9, 100]

# this creates 5 bins as follows,
# (0,59.9], (59.9,69.9], (69.9,79.9], (79.9,89.9], (89.9,100]

# Create the names for the five bins
group_names = ["F", "D", "C", "B", "A"]
```

```python
df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names)
df
```

```python
df2["Test Score Summary"] = pd.cut(df2["Test Score"], bins, labels=group_names, include_lowest=True)
df2
```

```python
# Creating a group based off of the bins
df = df.groupby("Test Score Summary")
df.mean()
```
